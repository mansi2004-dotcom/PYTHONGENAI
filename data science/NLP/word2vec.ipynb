```python
def cleanText(text):
    text = text.lower()
    # parenthesis
    text = re.sub(r'\([^)]*\)' , '', text)
    # \( -- opening parenthesis
    # \) -- closing parenthesis
    # [^)]* - ^ except this(jehra v essto baad character likhya hoyyega ohnu shadke) -- * zero or more 

    # white multi-white space with single space
    text = re.sub(r'\s+',' ', text)
    # r'' raw text
    #\s any white space character +(one or more)

    text = re.sub(r'["" '']','',text)
    text = text.strip()
    return text
```
