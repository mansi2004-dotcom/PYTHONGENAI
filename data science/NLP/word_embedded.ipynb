```python
sent = ['the glass of milk',
        'the glass of juice',
        'he is a good boy',
        'the cup of tea',
        'he is good developer',
        'you are very smart'
       ]
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
vocab_size =50
```

```python
one_hot_sent = [one_hot(i,vocab_size)for i in sent]
one_hot_sent
```

```python
# word embedding -- this layer is responsible for converting wordlayer to vector  
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
```

```python
sent_length =7
```

```python
embedded_docs = pad_sequences(one_hot_sent, maxlen=sent_length, padding='pre') # padding - post, pre
embedded_docs
```

```python
model = Sequential()
model.add(Embedding(vocab_size,5,input_length =sent_length))
model.compile('adam','mse')
```

```python
model.predict(embedded_docs)
```
