```python
import pandas as pd
import numpy as np
import sklearn
```

```python
data = pd.DataFrame({
    'Area': [1000, 1500, 2000],
    'Bedrooms': [2, 3, 3],
    'Location': ['NY', 'LA', 'Chicago']
})

data_encoded = pd.get_dummies(data, drop_first=True)
print(data_encoded.astype(int))
```

```python
data = pd.DataFrame({
    'Area': [1000, 1500, 2000],
    'Bedrooms': [2, 3, 3],
    'Location': ['NY', 'LA', 'Chicago']
})

# Get dummies
data_encoded = pd.get_dummies(data, drop_first=True)

# Convert boolean to integer
data_encoded = data_encoded.astype(int)

print(data_encoded)
```
