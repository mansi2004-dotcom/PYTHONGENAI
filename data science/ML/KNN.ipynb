```python
import pandas as pd
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = [[1,2],[2,3],[3,3],[6,7],[7,8]]
y = [10,30,40,50,90]
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)
print(model.predict([[5,6]]))
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
size=np.array([1000,1500,2000])
price=np.array([50000,70000,90000])
model = KNeighborsClassifier(n_neighbors=3)
model.fit(size,price)
print(model.predict([[size,price]]))
```
