```python
# Label Encoder is used to automatically set numerical value to categorical data . 
#Write Python code to convert categorical labels into numerical values using LabelEncoder.
from sklearn.preprocessing import LabelEncoder 
data1=['Cat','Eagle','Dog','Bird']
encoder = LabelEncoder()
encoded = encoder.fit_transform(data1)
print(encoded)
```

```python
#Write code to convert encoded values back to original labels.
decoded = encoder.inverse_transform(encoded)
print(decoded)
```

```python
#Apply LabelEncoder on a dataset column using pandas.
from sklearn.preprocessing import LabelEncoder 
data2=pd.DataFrame({'Fruits':['Apple','Orange','Banana','Peach','Grapes','Watermelon']})
df = pd.DataFrame(data2)
encoder1= LabelEncoder()
df['fruits_encoded']= encoder1.fit_transform(df['Fruits'])
print(df)
```

```python
#Find the unique classes learned by LabelEncoder.
from sklearn.preprocessing import LabelEncoder 
data3=['Red','Blue','Pink','Blue']
encoder_1= LabelEncoder()
encoder_1.fit(data3)
print(encoder_1.classes_)
```
