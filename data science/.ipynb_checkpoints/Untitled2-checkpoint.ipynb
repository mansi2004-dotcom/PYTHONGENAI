```python
import seaborn as sns
```

```python
fruits=['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]
plt.bar(x=fruits, height= sales, width=0.6)# x-axis - categories, y-axis - numerical values
plt.show()
```

```python
sns.barplot(x= sales , y = fruits)
```

```python
plt.bar(x=fruits, height=sales, width = 0.4, color = ['red','yellow','green','blue'])
```

```python
plt.bar(x=fruits, height=sales, width = 0.4, color = ['red','yellow','green','blue'])
plt.xlabel('fruits')
plt.ylabel('sales')
plt.title('fruits and sales')
plt.show()
```

```python
plt.barh(y=fruits, height=0.6, width = [120,150,190,200], color = ['red','yellow','green','blue'])
plt.xlabel('sales')
plt.ylabel('fruits')
plt.title('fruits and sales')
plt.show()
```
