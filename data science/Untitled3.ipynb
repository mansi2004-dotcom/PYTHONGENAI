```python
from sklearn.preprocessing import OneHotEncoder
```

```python
import pandas as pd
import seaborn as sns
data=sns.load_dataset('titanic')
data.head()
```

```python
data.who.value_counts()
```

```python
onehot= OneHotEncoder()
ans=onehot.fit_transform(data.loc[:,['who']]).toarray()
```

```python
ans
```

```python
new_features = onehot.get_feature_names_out()
new_features
```

```python
data.loc[:,new_features]=ans
```
