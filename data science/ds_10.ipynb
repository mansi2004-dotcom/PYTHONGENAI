```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.loc[:,'class'].value_counts()
```

```python
titanic.loc[:,'sex'].value_counts()
```

```python
sns.countplot(data=titanic, x='class' )
```

```python
sns.countplot(data=titanic, x='class', hue='survived')
```

```python
sns.countplot(data=titanic, x='class', hue='alive', order=['Third','Second','First'],hue_order=['yes','no'],width=0.6,palette='plasma') #fill true by default.
```
