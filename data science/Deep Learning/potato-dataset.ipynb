```python
import os
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Flatten, Dropout, MaxPooling2D
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1./255, zoom_range=0.2)
#20%
val_datagen = ImageDataGenerator(rescale=1./255)
# pixel value varies from 0 to 255
#neural networks better work on the small value data(0 to 1)
# every pixel is divided by 255
# pixel value is 128 = 128/255 => 0.8(approx..)
```

```python
Image_Size = 256
Batch_Size = 32
Channels = 3
Epochs = 40
```

```python
train_generator = train_datagen.flow_from_directory(
    '/kaggle/input/notebooks/priyark/potato-leaf-disease-97-100-accuracy',
    target_size=(256,256),  # 
    batch_size=32,  # Instead of sending all images at once, it sends 32 images at a time.
    class_mode='categorical'
)

# Every image will be resized to 256 Г— 256.
# Why?
# Because neural networks need fixed-size input.
# Even if original images are different sizes в†’ all become 256Г—256.
```

```python
classes =train_generator.class_indices  
classes
```

```python
classes = {j:i for i,j in classes.items()}
classes
```

```python
val_generator = val_datagen.flow_from_directory(
    '/kaggle/input/notebooks/priyark/potato-leaf-disease-97-100-accuracy',
    target_size=(256,256),
    batch_size=32,
    class_mode='categorical'
)
```

```python
from tensorflow.keras import datasets, layers, models
```

```python
# model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(256,256,3)),
    layers.MaxPooling2D((2,2)), 

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(4, activation='softmax') 
])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',  
    metrics=['accuracy']
)
```

```python
history = model.fit(
train_generator, epochs = Epochs, batch_size = Batch_Size, verbose = 1, validation_data = val_generator)
```
